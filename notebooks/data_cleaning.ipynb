```python
#cleaning trail spaces for name, artist, and album
#lowercasing all mood labels to avoid future SQL problems
df['name'] = df['name'].str.strip()
df['artist'] = df['artist'].str.strip()
df['album'] = df['album'].str.strip()
df['mood'] = df['mood'].str.strip()
```

```python
df_cleaned = pd.read_csv('../data/cleaned_songs.csv')
df_cleaned.head(10)
```

```python
import sqlite3
conn = sqlite3.connect('../data/songs.db')
cursor = conn.cursor()
```

```python
#columns: id, name, album, artist, release_date, popularity, length, danceability
#acousticness, energy, instrumentalness, liveness, valence, loudness, speechiness, 
#tempo, key, time_signature, mood
cursor.execute("""
CREATE TABLE songs (
    id TEXT PRIMARY KEY,
    name TEXT,
    album TEXT,
    artist TEXT,
    release_date TEXT,
    popularity INTEGER,
    length INTEGER,
    danceability REAL,
    acousticness REAL,
    energy REAL,
    instrumentalness REAL,
    liveness REAL,
    valence REAL,
    loudness REAL,
    speechiness REAL,
    tempo REAL,
    key INTEGER,
    time_signature INTEGER,
    mood TEXT)
""")
conn.commit()
```

```python
df_cleaned.to_sql('songs', conn, if_exists='append', index=False)
```

```python
cursor.execute("SELECT COUNT(*) FROM songs;")
rows = cursor.fetchone()
rows
```

```python
cursor.execute("PRAGMA table_info(songs);")
table_info = cursor.fetchall()
table_info

print("Column Names for 'songs' table:")
for column in table_info:
    print(column[1])
```
